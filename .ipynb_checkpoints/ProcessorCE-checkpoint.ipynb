```python
def load_and_melt(file_path, value_name):
    """
    åŠ è½½CSVæ–‡ä»¶å¹¶æ£€æŸ¥å¹´ä»½åˆ—ä¸­çš„æ— æ•ˆå€¼
    
    å‚æ•°:
        file_path (str): CSVæ–‡ä»¶è·¯å¾„
        value_name (str): è¦åˆ›å»ºçš„æŒ‡æ ‡åç§°
    
    è¿”å›:
        DataFrame: å¤„ç†åçš„æ•°æ®æ¡†
    """
    # è¯»å–CSVæ–‡ä»¶
    df = pd.read_csv(file_path)
    
    # æ¸…ç†æœªå‘½ååˆ—
    df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
    
    # è½¬æ¢ä¸ºé•¿æ ¼å¼
    melted = df.melt(
        id_vars='Countries',
        var_name='Year',
        value_name=value_name
    )
    
    # å°è¯•å°†å¹´ä»½è½¬æ¢ä¸ºæ•°å€¼ï¼Œæ— æ•ˆå€¼è®¾ä¸ºNaN
    melted['Year_numeric'] = pd.to_numeric(melted['Year'], errors='coerce')
    
    # æŸ¥æ‰¾æ— æ•ˆå¹´ä»½å€¼
    invalid_mask = melted['Year_numeric'].isna()
    invalid_rows = melted[invalid_mask]
    
    # å¦‚æœæœ‰æ— æ•ˆå€¼ï¼Œæ‰“å°è¯¦ç»†ä¿¡æ¯
    if not invalid_rows.empty:
        print(f"\nâš ï¸ åœ¨æ–‡ä»¶ {file_path} ä¸­å‘ç°æ— æ•ˆå¹´ä»½å€¼:")
        for idx, row in invalid_rows.iterrows():
            print(f"  è¡Œç´¢å¼•: {idx}, å›½å®¶: '{row['Countries']}', æ— æ•ˆå¹´ä»½å€¼: '{row['Year']}'")
        print(f"å…±å‘ç° {len(invalid_rows)} ä¸ªæ— æ•ˆå¹´ä»½å€¼\n")
    
    # åˆ é™¤æ— æ•ˆè¡Œå¹¶è½¬æ¢å¹´ä»½ä¸ºæ•´æ•°
    melted = melted.dropna(subset=['Year_numeric'])
    melted['Year'] = melted['Year_numeric'].astype(int)
    melted = melted.drop(columns=['Year_numeric'])
    
    return melted

'''
def load_and_melt(file_path, value_name):
    """
    è¯»å–æ–‡ä»¶å¹¶è½¬æ¢ä¸ºé•¿æ ¼å¼ï¼ˆCountry, Year, Valueï¼‰
    """
    # è¯»å–æ–‡ä»¶ï¼Œç¬¬ä¸€åˆ—ä¸ºå›½å®¶åç§°
    df = pd.read_csv(file_path)   
    # è½¬æ¢ä¸ºé•¿æ ¼å¼
    melted = df.melt(
        id_vars='Countries',
        var_name='Year',
        value_name=value_name
    )
    melted['Year'] = melted['Year'].astype(int)  # ç¡®ä¿å¹´ä»½ä¸ºæ•´æ•°
    return melted
'''
# ç¤ºä¾‹ï¼šåŠ è½½ CE æ•°æ®
ce_df = load_and_melt('Data1/processed/CO2EmissionsFromEnergy.csv', value_name='CE')
ce_df
```

### 2.1 è®¡ç®—å› ç´ 

```python
# æ•°æ®æ•´ç†
# å®šä¹‰åˆ†è§£æ‰€éœ€çš„å› å­åˆ—å
factors = [
    'total_CE/energy_consumption',
    'energy_consumption/GDP',
    'GDP/Population',
    'Population/cement_consumption(volume)',
    'cement_consumption(volume)/total_CS'
]

# åœ¨è®¡ç®—åˆ†è§£å› å­å‰æ·»åŠ å•ä½è½¬æ¢
print("æ‰§è¡Œå•ä½è½¬æ¢...")
# è´¨é‡å•ä½è½¬æ¢ (æ°´æ³¥äº§é‡: kt â†’ Mt) ç™¾ä¸‡å¨
merged_df['cement_production'] = merged_df['cement_production'] / 1000
# ç»æµå•ä½è½¬æ¢ (GDP: USD â†’ trillion USD) ä¸‡äº¿ç¾å…ƒ
merged_df['GDP'] = merged_df['GDP'] / 1e12
# äººå£å•ä½è½¬æ¢ (persons â†’ billion persons) äººå£(äº¿äºº)
merged_df['Population'] = merged_df['Population'] / 1e8
# ç°åœ¨è®¡ç®—åˆ†è§£å› å­ï¼ˆä½¿ç”¨ç»Ÿä¸€å•ä½ï¼‰
print("è®¡ç®—åˆ†è§£å› å­...")

# è®¡ç®—å„åˆ†è§£å› å­
merged_df['total_CE/energy_consumption'] = merged_df['total_CE'] / merged_df['energy_consumption']
merged_df['energy_consumption/GDP'] = merged_df['energy_consumption'] / merged_df['GDP']
merged_df['GDP/Population'] = merged_df['GDP'] / merged_df['Population']
merged_df['Population/cement_consumption(volume)'] = merged_df['Population'] / merged_df['cement_consumption(volume)']
merged_df['cement_consumption(volume)/total_CS'] = merged_df['cement_consumption(volume)'] / merged_df['cement_CS']

# ç¡®ä¿æ‰€æœ‰å…³é”®åˆ—å‡ä¸ºæ­£æ•°ï¼ˆé¿å…å¯¹éæ­£å€¼å–å¯¹æ•°ï¼‰
merged_df = merged_df[
    (merged_df['total_CE'] > 0) &
    (merged_df['cement_CS'] > 0) &
    (merged_df['cement_consumption(volume)'] > 0) &
    (merged_df['cement_production'] > 0) &
    (merged_df['GDP'] > 0) &
    (merged_df['Population'] > 0)&
    (merged_df['energy_consumption'] > 0) 
]

# å†æ¬¡æ£€æŸ¥æ— ç©·å¤§å’Œç¼ºå¤±å€¼
# print("æ¸…ç†æ— ç©·å€¼å’Œç¼ºå¤±å€¼...")
# merged_df = merged_df.replace([np.inf, -np.inf], np.nan)
merged_df = merged_df.dropna(subset=factors + ['total_CE', 'cement_CS'])
print(f"æœ€ç»ˆæœ‰æ•ˆæ•°æ®é‡: {len(merged_df)} è¡Œ (1975å¹´ä¹‹å)")
print(f"æœ€ç»ˆæœ‰æ•ˆå›½å®¶å…±{len(merged_df["Countries"].unique())}ä¸ªï¼Œåˆ—è¡¨: {merged_df["Countries"].unique()}")
merged_df[merged_df["Countries"] == "China"]
```

## 3.LMDIåˆ†è§£

```python
def lmdi_decomposition(df, base_year, target_ratio='total_CE/cement_CS'):
    
    """
    æ‰§è¡Œ LMDI åˆ†è§£ï¼Œè®¡ç®—å„å› å­è´¡çŒ®
    - df: åˆå¹¶åçš„æ•°æ®ï¼ˆåŒ…å« Country, Year å’Œå„å› å­ï¼‰
    - base_year: åŸºæœŸå¹´ä»½ï¼ˆå¦‚ 1990ï¼‰
    - target_ratio: åˆ†è§£çš„ç›®æ ‡æ¯”ä¾‹ï¼ˆé»˜è®¤ä¸º CS/CEï¼‰
    """
    
    # è®¡ç®—ç›®æ ‡æ¯”ä¾‹ CFP = CE/CS
    df[target_ratio] = df['total_CE'] / df['cement_CS']
    

    # åˆå§‹åŒ–ç»“æœå­˜å‚¨
    results = []
    for country, country_df in df.groupby('Countries'):
        country_df = country_df.sort_values('Year')
        country_df = country_df[country_df['Year'] >= base_year]
        if len(country_df) < 2:
            continue

        for idx in range(1, len(country_df)):
            t_row = country_df.iloc[idx]
            b_row = country_df.iloc[idx-1]

            cfp_t = t_row[target_ratio]
            cfp_b = b_row[target_ratio]

            # æ·»åŠ å¯¹ CFP å€¼çš„ä¿æŠ¤
            if cfp_t <= 0 or cfp_b <= 0:
                continue  # è·³è¿‡éæ­£å€¼

            # è®¡ç®— L(CFP^t, CFP^b)
            if np.isclose(cfp_t, cfp_b, rtol=1e-6):  # æµ®ç‚¹æ•°ç²¾åº¦å®¹é”™
                L = cfp_t
            else:
                log_diff = np.log(cfp_t) - np.log(cfp_b)
                '''
                # é˜²æ­¢é™¤ä»¥é›¶
                if np.isclose(log_diff, 0):  
                    L = cfp_t
                else:
                    L = (cfp_t - cfp_b) / log_diff
                '''
                L = (cfp_t - cfp_b) / log_diff

            # è®¡ç®—å„å› å­è´¡çŒ®ï¼ˆæ·»åŠ å¯¹ ratio_t/ratio_b çš„ä¿æŠ¤ï¼‰
            contributions = {}
            for factor in factors:
                ratio_t = t_row[factor]
                ratio_b = b_row[factor]
                if ratio_t <= 0 or ratio_b <= 0:
                    contributions[f'Î”{factor}'] = np.nan
                else:
                    contributions[f'Î”{factor}'] = L * np.log(ratio_t / ratio_b)

            '''
            # è·³è¿‡å­˜åœ¨ NaN çš„è¡Œ
            if not np.isnan(list(contributions.values())).any():
                results.append({
                    'Countries': country,
                    'Year': t_row['Year'],
                    'Î”Total': cfp_t - cfp_b,
                    **contributions
                })
            '''

            results.append({
                    'Countries': country,
                    'Year': t_row['Year'],
                    'Î”Total': cfp_t - cfp_b,
                    **contributions
                })
        
    return pd.DataFrame(results)
```

```python
# è®¾ç½®åŸºæœŸå¹´ä»½ï¼ˆæ ¹æ®æ•°æ®å®é™…èµ·å§‹æ—¶é—´è°ƒæ•´ï¼‰
base_year = 1975  # ç¤ºä¾‹

# è¿è¡Œåˆ†è§£
lmdi_results = lmdi_decomposition(merged_df, base_year)

# ä¿å­˜ç»“æœ
lmdi_results.to_csv('Data1/results/LMDI_Contributions.csv', index=False, encoding='utf-8-sig')

# æŸ¥çœ‹ç¤ºä¾‹è¾“å‡º
print(lmdi_results.head())
```

```python
# éªŒè¯ï¼š
lmdi_results['Î”Sum'] = lmdi_results[[f'Î”{factor}' for factor in factors]].sum(axis=1)
print(lmdi_results[['Î”Total', 'Î”Sum']].head())
```

```python
def lmdi_decomposition_single_period(df, base_year, target_ratio='cement_CS/total_CE'):
    """
    æ‰§è¡ŒLMDIåˆ†è§£ï¼ˆæ¯ä¸ªå›½å®¶åªè®¡ç®—æœ€æ–°ä¸€å¹´ä¸æœ€å¼€å§‹ä¸€å¹´çš„å¯¹æ¯”ï¼‰
    - df: åˆå¹¶åçš„æ•°æ®ï¼ˆåŒ…å«Country, Yearå’Œå„å› å­ï¼‰
    - base_year: åŸºæœŸå¹´ä»½ï¼ˆå¦‚1975ï¼‰
    - target_ratio: åˆ†è§£çš„ç›®æ ‡æ¯”ä¾‹ï¼ˆé»˜è®¤ä¸ºCE/CSï¼‰
    """ 
    # è®¡ç®—ç›®æ ‡æ¯”ä¾‹ CFP = CE/CS
    df[target_ratio] = df['total_CE'] / df['cement_CS']
    
    # åˆå§‹åŒ–ç»“æœå­˜å‚¨
    results = []
    for country, country_df in df.groupby('Countries'):
        # ç­›é€‰åŸºæœŸä¹‹åçš„æ•°æ®å¹¶æŒ‰å¹´ä»½æ’åº
        country_df = country_df[country_df['Year'] >= base_year].sort_values('Year')
        # if len(country_df) < 2:
        #    continue  # è‡³å°‘éœ€è¦é¦–å°¾ä¸¤å¹´æ•°æ®
        
        # è·å–é¦–å°¾ä¸¤å¹´æ•°æ®
        first_row = country_df.iloc[0]    # æœ€å¼€å§‹ä¸€å¹´ï¼ˆ>= base_yearçš„æœ€æ—©å¹´ä»½ï¼‰
        last_row = country_df.iloc[-1]    # æœ€æ–°ä¸€å¹´
        
        cfp_first = first_row[target_ratio]
        cfp_last = last_row[target_ratio]

        # è·³è¿‡éæ­£å€¼
        # if cfp_first <= 0 or cfp_last <= 0:
        #    continue

        # è®¡ç®—L(CFP_last, CFP_first)
        if np.isclose(cfp_last, cfp_first, rtol=1e-6):
            L = cfp_last
        else:
            log_diff = np.log(cfp_last) - np.log(cfp_first)
            L = (cfp_last - cfp_first) / log_diff if not np.isclose(log_diff, 0) else cfp_last

        # è®¡ç®—å„å› å­è´¡çŒ®
        contributions = {}
        for factor in factors:
            ratio_first = first_row[factor]
            ratio_last = last_row[factor]
            if ratio_first <= 0 or ratio_last <= 0:
                contributions[f'Î”{factor}'] = np.nan
            else:
                contributions[f'Î”{factor}'] = L * np.log(ratio_last / ratio_first)
        
        # ä¿å­˜ç»“æœï¼ˆæ¯ä¸ªå›½å®¶ä¸€æ¡è®°å½•ï¼‰
        if not np.isnan(list(contributions.values())).any():
            results.append({
                'Countries': country,
                'Start_Year': first_row['Year'],
                'End_Year': last_row['Year'],
                'Î”Total': cfp_last - cfp_first,  # æ€»å˜åŒ–é‡
                **contributions
            })
    
    return pd.DataFrame(results)
```

```python
# è¿è¡Œåˆ†è§£ï¼ˆæ¯ä¸ªå›½å®¶è¾“å‡ºé¦–å°¾å¹´å¯¹æ¯”ï¼‰
lmdi_results = lmdi_decomposition_single_period(merged_df, base_year=1975)

# ä¿å­˜ç»“æœ
lmdi_results.to_csv('Data1/results/CE_LMDI_Contributions_20years.csv', index=False)

# æŸ¥çœ‹ç¤ºä¾‹è¾“å‡º
print(lmdi_results.head())
```

### 5 å¹´æ—¶é—´é—´éš”åˆ†è§£

```python
import numpy as np
import pandas as pd

def lmdi_decomposition_5(df, base_year, target_ratio='total_CE/cement_CS'):
    """
    æ‰§è¡Œæ¯5å¹´ä¸€æ¬¡çš„LMDIåˆ†è§£ï¼Œè®¡ç®—å„å› å­è´¡çŒ®
    - df: åŒ…å« Countries, Year å’Œå„å› å­çš„æ•°æ®
    - base_year: åŸºæœŸèµ·å§‹å¹´ä»½
    - factors: åˆ†è§£çš„å› å­åˆ—è¡¨
    - target_ratio: åˆ†è§£çš„ç›®æ ‡æ¯”ä¾‹
    """
    
    # è®¡ç®—ç›®æ ‡æ¯”ä¾‹ CFP = CE/CS
    df[target_ratio] = df['total_CE'] / df['cement_CS']
    
    results = []
    for country, country_df in df.groupby('Countries'):
        country_df = country_df.sort_values('Year')
        country_df = country_df[country_df['Year'] >= base_year].copy()
        
        if len(country_df) < 1:
            continue
            
        # åˆ›å»º5å¹´åˆ†ç»„
        country_df['period'] = (country_df['Year'] - base_year) // 5
        
        for period, period_df in country_df.groupby('period'):
            period_df = period_df.sort_values('Year')
            start_year = period_df['Year'].min()
            end_year = period_df['Year'].max()+1
            
            # è·³è¿‡åªæœ‰ä¸€ä¸ªå¹´ä»½çš„å‘¨æœŸï¼ˆæ— æ³•è®¡ç®—å˜åŒ–ï¼‰
            if len(period_df) < 2:
                continue
                
            # å–å‘¨æœŸå¼€å§‹å¹´å’Œç»“æŸå¹´çš„æ•°æ®
            start_row = period_df.iloc[0]
            end_row = period_df.iloc[-1]
            
            cfp_t = end_row[target_ratio]
            cfp_b = start_row[target_ratio]
            
            # è·³è¿‡æ— æ•ˆå€¼
            if cfp_t <= 0 or cfp_b <= 0:
                continue
                
            # è®¡ç®—L(CFP^t, CFP^b)
            if np.isclose(cfp_t, cfp_b, rtol=1e-6):
                L = cfp_t
            else:
                log_diff = np.log(cfp_t) - np.log(cfp_b)
                L = (cfp_t - cfp_b) / log_diff
                
            # è®¡ç®—å„å› å­è´¡çŒ®
            contributions = {}
            for factor in factors:
                ratio_t = end_row[factor]
                ratio_b = start_row[factor]
                
                if ratio_t <= 0 or ratio_b <= 0:
                    contributions[f'Î”{factor}'] = np.nan
                else:
                    contributions[f'Î”{factor}'] = L * np.log(ratio_t / ratio_b)
            
            results.append({
                'Countries': country,
                'Period': f"{start_row['Year']}-{end_row['Year']+1}",
                'Start_Year': start_year,
                'End_Year': end_year,
                # 'Years': end_year - start_year + 1,
                'Î”Total': cfp_t - cfp_b,
                **contributions
            })
            
    return pd.DataFrame(results)
```

## 4.è®¡ç®—å•ä½CSå˜åŒ–äº§ç”Ÿçš„è´¡çŒ®å˜åŒ–ç‡
### æ ¹æ®tapioè§£è€¦æ€è·¯ï¼Œè®¡ç®—å› ç´ å˜åŒ–é‡ï¼Œç”¨æ¥è¡¡é‡å•ä½å› ç´ å˜åŒ–çš„è´¡çŒ®å˜åŒ–é‡

$ \frac{(C_5^T - C_5^0) / \Delta C_5^0}{(CS^T - CS^0)/CS^0} $

### 4.1 è®¡ç®—5å¹´å› ç´ ï¼ˆCSï¼‰å˜åŒ–ç‡

```python
def calc_factor_delta_5(df, base_year, factors):
    """
    æ‰§è¡Œæ¯5å¹´ä¸€æ¬¡çš„å› å­å˜åŒ–ç‡è®¡ç®—
    - df: åˆå¹¶åçš„æ•°æ®ï¼ˆåŒ…å« Country, Year å’Œå„å› å­ï¼‰
    - base_year: åŸºæœŸå¹´ä»½ï¼ˆå¦‚ 1990ï¼‰
    - factors: è¦è®¡ç®—çš„å› å­åˆ—è¡¨
    """
    # åˆå§‹åŒ–ç»“æœå­˜å‚¨
    results = []
    for country, country_df in df.groupby('Countries'):
        country_df = country_df.sort_values('Year')
        country_df = country_df[country_df['Year'] >= base_year].copy()
        
        if len(country_df) < 1:
            continue
            
        # åˆ›å»º5å¹´åˆ†ç»„
        country_df['period'] = (country_df['Year'] - base_year) // 5
        
        for period, period_df in country_df.groupby('period'):
            period_df = period_df.sort_values('Year')
            
            # è·³è¿‡åªæœ‰ä¸€ä¸ªå¹´ä»½çš„å‘¨æœŸï¼ˆæ— æ³•è®¡ç®—å˜åŒ–ï¼‰
            if len(period_df) < 2:
                continue
                
            # å–å‘¨æœŸå¼€å§‹å¹´å’Œç»“æŸå¹´çš„æ•°æ®
            start_row = period_df.iloc[0]
            end_row = period_df.iloc[-1]
            
            # è®¡ç®—å„è´¡çŒ®å¯¹åŸºæœŸå˜åŒ–ç‡
            gradients = {}
            for factor in factors:
                ratio_t = end_row[factor]
                ratio_b = start_row[factor]
                
                # å¤„ç†åˆ†æ¯ä¸º0æˆ–è´Ÿå€¼çš„æƒ…å†µ
                if ratio_b <= 0 or pd.isna(ratio_b):
                    gradients[f'Î”{factor}'] = np.nan
                else:
                    gradients[f'Î”{factor}'] = (ratio_t - ratio_b) / ratio_b
            
            results.append({
                'Countries': country,
                'Period': f"{start_row['Year']}-{end_row['Year']+1}",
                # 'Start_Year': start_row['Year'],
                # 'End_Year': end_row['Year'],
                # 'Years': end_row['Year'] - start_row['Year'],
                **gradients
            })
    
    return pd.DataFrame(results)

factors = [
    'cement_CS'
]
factor_delta_5 = calc_factor_delta_5(merged_df, 1975, factors) 
factor_delta_5
```

### è®¡ç®—5å¹´åŒºé—´è´¡çŒ®å˜åŒ–ç‡

```python
def calc_contribution_delta_5(df, base_year, factors):
    """
    æ‰§è¡Œ Tapio è§£è€¦ï¼Œè®¡ç®—å„å› å­å˜åŒ–ç‡
    - df: åˆå¹¶åçš„æ•°æ®ï¼ˆåŒ…å« Country, Year å’Œå„å› å­ï¼‰
    - base_year: åŸºæœŸå¹´ä»½ï¼ˆå¦‚ 1990ï¼‰
    """
    # åˆå§‹åŒ–ç»“æœå­˜å‚¨
    results = []
    for country, country_df in df.groupby('Countries'):
        country_df = country_df.sort_values('Start_Year')
        country_df = country_df[country_df['Start_Year'] >= base_year]
        if len(country_df) < 2:
            continue
            
        for idx in range(1, len(country_df)):
            t_row = country_df.iloc[idx]
            b_row = country_df.iloc[idx-1]
    
            # è®¡ç®—å„å› å­å˜åŒ–ç‡
            gradients = {}
            for factor in factors:
                ratio_t = t_row[factor]
                ratio_b = b_row[factor]
                if ratio_b <= 0:
                    gradients[f'Î”{factor}'] = np.nan
                else:
                    gradients[f'Î”{factor}'] = (ratio_t-ratio_b)/ratio_b
    
            results.append({
                    'Countries': country,
                    'Period': str(b_row['Start_Year']) + '-' + str(t_row['Start_Year']),
                    **gradients
                })
    return pd.DataFrame(results)

factors = [
    'Î”total_CE/energy_consumption',
    'Î”energy_consumption/GDP',
    'Î”GDP/Population',
    'Î”Population/cement_consumption(volume)',
    'Î”cement_consumption(volume)/total_CS'
]
contribution_delta_5 = calc_contribution_delta_5(lmdi_results_5, 1975, factors) 
contribution_delta_5
```

```python
rt_df = pd.merge(
            contribution_delta_5,
            factor_delta_5, 
            on=['Countries','Period'],  # æ³¨æ„åˆ—åæ˜¯'Countries'è€Œä¸æ˜¯'Country'
            how='inner'
        )
rt_df = rt_df[["Countries", "Period","Î”Î”cement_consumption(volume)/total_CS","Î”cement_CS"]]
rt_df
```
